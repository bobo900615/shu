```python
# 10
import random
year1=eval(input("enter year1:"))
year2=eval(input("enter year2:"))
for i in range (year1,year2):
    randomnumber=random.randint(year1,year2)
    if i %10==0:
        print("%5d"%(randomnumber))
    else:
        print("%5d"%(randomnumber),end="")
        if randomnumber %4==0:
            print(randomnumber,"是閏年")
        else:
            print(randomnumber,"不是閏年")
```
